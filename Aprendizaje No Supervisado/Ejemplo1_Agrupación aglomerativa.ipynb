## Ejemplo de Agglomerative Clustering

## Con Dataset aleatorio

```python
# Generar clusters aleatorios
from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt

# Generar datos de ejemplo
X, labels = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std=4, random_state=123)

# Visualizar los datos
plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.7)
plt.title("Datos de ejemplo con 2 clusters")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

```python
# Imprimir los 10 primeros puntos de datos 
print(X[:10])
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd

# Aplicar agrupacin aglomerativa
agglo = AgglomerativeClustering(n_clusters=2, linkage='average') # linkage puede ser 'ward', 'complete', 'average', etc.
agglo.fit(X) # Entrenar el modelo

# Obtener las etiquetas de los clusters
labels = agglo.labels_

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.7)
plt.title("Resultados de la agrupacin aglomerativa")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

### Visualizacin en Dendrograma

```python
# Dendrogram --> Visualizacin en Dendrograma
# linkage => ejecutar el algoritmo y representar las uniones
# fcluster => cortar a nivel especfico y ver los clusters resultantes
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Calcular la matriz de enlace
Z = linkage(X, method='average')

# Visualizar el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title("Dendrograma de la agrupacin aglomerativa")
plt.xlabel("ndice de muestra")
plt.axhline(y=13, color='r', linestyle='--')  # Lnea de corte
plt.ylabel("Distancia")
plt.show()
```

```python
# etiquetas de clusters a partir del dendrograma
labels = fcluster(Z, t=13, criterion='distance') # t es el umbral de corte
pd.Series(labels).value_counts()
```

## Con otro Datasets

```python
from sklearn.datasets import make_moons
# Crear un dataset de lunas
X_moons, labels_moons = make_moons(n_samples=200, noise=0.15, random_state=567)  

# Visualizar el dataset de lunas
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=labels_moons, cmap='plasma', alpha=0.7)
plt.title("Dataset de lunas")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd

# Aplicar agrupacin aglomerativa
agglo = AgglomerativeClustering(n_clusters=2, linkage='ward') # linkage puede ser 'ward', 'complete', 'average', etc.
agglo.fit(X_moons) # Entrenar el modelo

# Obtener las etiquetas de los clusters
labels = agglo.labels_

# Visualizar los resultados
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=labels, cmap='plasma', alpha=0.7)
plt.title("Resultados de la agrupacin aglomerativa")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

```python
# Visualizacin en Dendrograma
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Calcular la matriz de enlace
Z = linkage(X_moons, method='complete')

# Visualizar el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title("Dendrograma de la agrupacin aglomerativa - Dataset de lunas")
plt.xlabel("ndice de muestra")
plt.axhline(y=3, color='r', linestyle='--')  # Lnea de corte
plt.ylabel("Distancia")
plt.show()
```

```python
# etiquetas de clusters a partir del dendrograma
labels = fcluster(Z, t=3, criterion='distance') # t es el umbral de corte
pd.Series(labels).value_counts()
```

## Datasets de Crculos concentrcos

```python
from sklearn.datasets import make_circles
# Crear un dataset de crculos concntricos
X_circles, labels_circles = make_circles(n_samples=200, noise=0.04, factor=0.5, random_state=567)

# Visualizar el dataset de crculos
plt.scatter(X_circles[:, 0], X_circles[:, 1], c=labels_circles, cmap='plasma', alpha=0.7)
plt.title("Dataset de crculos")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

```python
from sklearn.datasets import make_classification

# Crear un dataset de clasificacin
    # n_features --> Nmero de dimensiones
    # n_informative --> Nmero de caractersticas informativas
    # n_classes --> Nmero de clusters
X_classification, labels_classification = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=4, class_sep=2, random_state=53)

# Visualizar el dataset de clasificacin
plt.scatter(X_classification[:, 0], X_classification[:, 1], c=labels_classification, cmap='plasma', alpha=0.7)
plt.title("Dataset de clasificacin")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```

## Combinacin de Datasets

```python
# Combinacin de Datasets
x_moons, label_moons = make_moons(n_samples=100, noise=0.1, random_state=123)

# Creamos blob que ser el ruido de fondo
x_noise, label_noise = make_blobs(n_samples=100, centers=[(2, 2)], cluster_std=0.8, random_state=123)
label_noise[:] = 2  # Asignamos una etiqueta diferente para el ruido en amarillo


# Combinamos los datasets
X_combined = np.vstack([x_moons, x_noise])
y_combined = np.hstack([label_moons, label_noise])

# Visualizar los datos combinados
plt.scatter(X_combined[:, 0], X_combined[:, 1], c=y_combined, cmap='plasma', alpha=0.8)
plt.title("Datos combinados: lunas + ruido")
plt.xlabel("Caracterstica 1")
plt.ylabel("Caracterstica 2")
plt.show()
```
